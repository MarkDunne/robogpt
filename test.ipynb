```python
import os
import base64
from openai import AsyncAzureOpenAI
from dotenv import load_dotenv

load_dotenv()

API_VERSION = "2025-03-01-preview"
MODEL = "gpt-5"

client = AsyncAzureOpenAI(
    azure_endpoint=os.getenv("AZURE_OPENAI_ENDPOINT"),
    api_version=API_VERSION,
    api_key=os.getenv("AZURE_OPENAI_API_KEY"),
)
```

```python
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")


base64_image = encode_image("imgs/tiger.jpeg")


response = await client.responses.create(
    model=MODEL,
    input=[
        {
            "role": "user",
            "content": [
                {
                    "type": "input_image",
                    "image_url": f"data:image/jpeg;base64,{base64_image}",
                },
                {
                    "type": "input_text",
                    "text": "Analyse the above image, and tell me when you're ready. do not say anything about the content of the image just yet, just say you're ready. I'll ask you about the image later.",
                },
            ],
        }
    ],
)

print(response.output_text)

response = await client.responses.create(
    previous_response_id=response.id,
    model=MODEL,
    input=[
        {
            "role": "user",
            "content": [
                {
                    "type": "input_text",
                    "text": "Now tell me what is in the earlier image?",
                },
            ],
        }
    ],
)

print(response.output_text)
```
